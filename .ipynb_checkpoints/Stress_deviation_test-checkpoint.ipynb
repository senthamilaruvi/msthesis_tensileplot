```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
from scipy.optimize import leastsq
```

```python
root = os.chdir('E:\Python\Tensile_Plot\Data\A09')

# Reading all the data in csv format to the data frame

output = pd.read_csv("output.csv", skiprows = 3, names = ['Time','Force','Disp'])
eystrain = pd.read_csv("ey_strain.csv",skiprows = 3,names =['Stage','ey_strain'])
mises = pd.read_csv("mises.csv",skiprows = 3, names =['Stage','mises_strain'])
Dimensions = pd.read_table("Dimensions.txt")

# Dropping rows with missing values
output = output.dropna(axis = 0, subset =['Time', 'Force'])
eystrain = eystrain.dropna(axis =0 , subset =['ey_strain', 'Stage'])
mises = mises.dropna(axis = 0, subset =['mises_strain','Stage'])
print('Input_length:')
print(len(eystrain))
```

```python
#Calculating Stress from the output data 
Area = int(Dimensions["Width"]*Dimensions["Thickness"])


# Changing strain from percentage to mm/mm format
eystrain["ey_strain"] = eystrain["ey_strain"]/100
mises["mises_strain"] = mises["mises_strain"]/100

# Calculating Stress from force 
output["stress"] = output["Force"]/Area 

# Assembling all the data to a data frame
data = pd.concat(
    [output['Time'], output['stress'], eystrain['Stage'], eystrain['ey_strain'], mises['mises_strain']], 
    axis = 1, keys= ['time','stress','stage','strain','mises'])

print(data.head(10))
```

```python
#Filtering stress

#Filtering out the Stress data corresponding to each Strain Stage

Stress = []
time = np.array(data['time'])
stage = np.array(data['stage'].dropna())
time_sorted = []

# Finding closest time value of stress corresponding to each strain stage data
for i in range(0, len(stage)):
    time_sorted.append(min(range(len(time)), key=lambda j: abs(time[j]-stage[i])))

for i in range (0, len(time_sorted)):
        Stress.append(data['stress'][time_sorted[i]])
Stress = np.asarray(Stress)

Strain = np.array(data["strain"].dropna())
Mises = np.array(data["mises"].dropna())

print(len(Strain))
print(len(Stress))
```

```python
plt.figure(figsize=(16,9))
plt.plot(Strain, Stress)
plt.grid()
plt.xlim(0, 0.03)
```

```python
moving_average = lambda arr, window: np.array([np.mean(arr[i:i+window]) for i in xrange(len(arr))])
```

```python
epsilon, sigma = Strain[1:], Stress[1:]
modulus = lambda e,s: s/e
plt.figure(figsize=(16,9))
plt.plot(epsilon, modulus(epsilon, sigma))
for window in (2, 4, 8, 16):
    plt.plot(moving_average(epsilon, window), modulus(moving_average(epsilon, window), moving_average(sigma, window)),
             label='{:d}'.format(window))
plt.legend()
plt.axvline(0.007)
plt.xlim(0, 0.03)
```

```python
def strain_deviation(strain, stress, modulus):
    return strain - (stress/modulus)

mask = epsilon < 0.01
plt.figure(figsize=(16,9))
for trialE in range(78000, 82000, 500):
    plt.plot(strain_deviation(epsilon[mask], sigma[mask], trialE), sigma[mask], 'o', label='{:d}'.format(trialE))
plt.axvline(0)
plt.legend()
```

```python
mask = Strain < 0.0075
fit = stats.linregress(Strain[mask], Stress[mask])
```

```python
x = np.linspace(0, 0.02)
plt.figure(figsize=(16,9))
plt.plot(Strain, Stress)
plt.plot(x, fit.slope*x + fit.intercept)
plt.xlim(0, 0.02)
```
